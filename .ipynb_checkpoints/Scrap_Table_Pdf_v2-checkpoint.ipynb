```python
import tabula
import os
```

```python
# read PDF file
#tables = tabula.read_pdf("Dr. Waniss A. Otman_Libyan Oil and Gas Resources_First Edition_2008.pdf", pages="all")
```

```python
tables= cam.read_pdf("Dr. Waniss A. Otman_Libyan Oil and Gas Resources_First Edition_2008.pdf", 'wb')
tables
```

```python
# save them in a folder
folder_name = "tables"
if not os.path.isdir(folder_name):
    os.mkdir(folder_name)
```

```python
# iterate over extracted tables and export as excel individually
for i, table in enumerate(tables, start=1):
    table.to_excel(os.path.join(folder_name, f"table_{i}.xlsx"), index=False)
```
